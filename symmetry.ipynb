```python
from pymatgen import Structure
from pymatgen.ext.matproj import MPRester
from pymatgen.symmetry.analyzer import SpacegroupAnalyzer
from jupyter_jsmol.pymatgen import quick_view
```

```python
with MPRester("BQb2tXA4gWoqynx7") as mp:
    # Structure for material id
    structure = mp.get_structure_by_material_id("mp-984761")
```

```python
symmetry = SpacegroupAnalyzer(structure).get_symmetry_operations()
Cs = [0, 0, 1/2]
Rb = [1/4, 1/4, 1/4]
Sb = [0, 0, 0]
```

# Opération de symétrie sur l'atome de Césium

```python
Cs2 = symmetry[27].operate(Cs)
print("Type de symétrie : Plan miroir (z, y, x)")
print("Coordonnées du Césium avant la symétrie : ", Cs)
print("Coordonnées du Césium après la symétrie : ", Cs2)
print("\n")
print("Matrice de rotation : \n", symmetry[27].rotation_matrix)
print("\n")
print("Matrice de translation : \n", symmetry[27].translation_vector)
```

# Opération de symétrie sur l'atome de Rubidium

```python
Rb2 = symmetry[1].operate(Rb)
print("Type de symétrie : Inversion de centre (0, 0, 0)")
print("Coordonnées du Rubidium avant la symétrie : ", Rb)
print("Coordonnées du Rubidium après la symétrie : ", Rb2)
print("\n")
print("Matrice de rotation : \n", symmetry[1].rotation_matrix)
print("\n")
print("Matrice de translation : \n", symmetry[1].translation_vector)
```

# Opération de symétrie sur l'atome d'Antimoine

```python
Sb2 = symmetry[16].operate(Sb)
print("Type de symétrie : Rotation d'axe C3 (z, x, y)")
print("Coordonnées de l'Antimoine avant la symétrie : ", Sb)
print("Coordonnées de l'Antimoine après la symétrie : ", Sb2)
print("\n")
print("Matrice de rotation : \n", symmetry[16].rotation_matrix)
print("\n")
print("Matrice de translation : \n", symmetry[16].translation_vector)
```
