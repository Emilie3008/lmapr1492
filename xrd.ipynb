Ce Jupyter notebook se propose d'analyser les 3 premiers pics de diffraction obtenus lors du bombardement d'un rayonnement X incident Cu Kα (λ=1.54060 Angström) sur le cristal CsRb2Sb.

```python
from pymatgen.ext.matproj import MPRester
from pymatgen.analysis.diffraction.xrd import XRDCalculator
```

```python
with MPRester("BQb2tXA4gWoqynx7") as mp:
    structure = mp.get_structure_by_material_id("mp-984761")
```

# Diffractogramme par un rayonnement X Cu Kα (λ=1.54060 Å)

```python
xrd = XRDCalculator('CuKa',1.54060)
pattern = xrd.get_pattern(structure, two_theta_range=(0,35))
xrd.show_plot(structure,annotate_peaks=False, two_theta_range=(0,35))
```

# Interprétation du diffractogramme

Premier pic :

```python
theta = pattern.x[0]
intensity = pattern.y[0]
hkl = pattern.hkls[0]

print("Le premier pic se situe à 2θ = " + str(theta)+"° avec une intensité de " + str(intensity)+" et les indices hkl associés sont " +str(hkl[0]['hkl'])+" avec une multiplicité de "+str(hkl[0]['multiplicity']))
```
