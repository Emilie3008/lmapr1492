```python
from pymatgen.ext.matproj import MPRester
from pymatgen.symmetry.analyzer import SpacegroupAnalyzer
```

```python
with MPRester("BQb2tXA4gWoqynx7") as mp:
    # Structure for material id
    structure = mp.get_structure_by_material_id("mp-984761")
    structure.to(filename="CsRb2Sb.cif")

lat = structure.lattice
#récupération des paramètres de maille
a= lat.a; b=lat.b; c=lat.c
print("Réseau direct : \n")
print("a: (",a,", 0, 0) \n")
print("b: (0,",b,", 0) \n")
print("c: (0, 0, ",c,") \n")
#réseau réciproque
A = lat.reciprocal_lattice_crystallographic.a
B = lat.reciprocal_lattice_crystallographic.b
C = lat.reciprocal_lattice_crystallographic.c
print("Réseau réciproque : \n")
print("a: (",A,", 0, 0) \n")
print("b: (0,",B,", 0) \n")
print("c: (0, 0, ",C,") \n")
```

```python
analyser = SpacegroupAnalyzer(structure)
maille = analyser.get_lattice_type() #type de maille du cristal
systeme = analyser.get_crystal_system() #système cristallin
groupe = analyser.get_point_group_symbol() #groupe ponctuel
print("Type de maille du CsRb2Sb : ",maille," \n")
print("Système cristallin du CsRb2Sb : ",systeme," \n")
print("Groupe ponctuel de CsRb2Sb : ",groupe," \n")
```
